```python
df['Target'] = df['Target'].replace('#VALUE!', '0')
```

```python
df['Target'] = pd.to_numeric(df['Target'])
```

```python
for i in df.index:
    if df.loc[i,'Target'] == 0:
        df.loc[i , 'Target'] = df['Target'].mean()
```

```python
df[df['Delivery_person_Ratings'] < 2]
```

```python

for i in df.index:
    if df.loc[i,'Delivery_person_Ratings'] < 2:
        df.drop(i,inplace=True)

```

```python
df[df['Delivery_person_Ratings'] < 2]
```

```python
x = df['Temperature'].mean()

df['Temperature'] = df['Temperature'].fillna(x)
```

```python
df.isnull().sum()
```

```python
target  = df['Target'].mode()[0]

df['Target'] = df['Target'].fillna(target)

df['Target']
```

```python
df['Weather'] = df['Weather'].fillna(df['Weather'].mode()[0])
```

```python
df.isnull().sum()
```

```python
humidity = df['Humidity'].mean()

df['Humidity'] = df['Humidity'].fillna(humidity)
```

```python
df['Distance (km)'] = df['Distance (km)'].fillna(df['Distance (km)'].median())
```

```python
df.isnull().sum()
```

```python
df['Traffic']
```

```python
df['Traffic'] = df['Traffic'].fillna(df['Traffic'].mode()[0])
```

```python
df.isnull().sum()
```

```python
rl = df['Restaurant_latitude'].mode()[0]
rlong = df['Restaurant_longitude'].mode()[0]
```

```python
for i in df.index:
    if df.loc[i,'Restaurant_latitude'] == 0:
        df.loc[i , 'Restaurant_latitude'] = rl
    if df.loc[i,'Restaurant_longitude'] == 0:
        df.loc[i , 'Restaurant_longitude'] = rlong
```
